### 8. Write a function that checks if customer borrowed more or less films in the month of June as compared to May.

```python
## merge both dataframes to check the values in one table
## using outer join to gather all the entries from both tables

rentals_may_and_june = rentals_may.merge(rentals_june, how='outer', left_on='customer_id', right_on='customer_id', suffixes = ('_may', '_june'))
rentals_may_and_june
```

```python
## replaceing NANs for 0

rentals_may_and_june.update(rentals_may_and_june[['rentals_per_customer_may','rentals_per_customer_june']].fillna(0))
rentals_may_and_june
```

```python
# function to check if the customers as a whole borrowed more films in june comparing with may

if rentals_may_and_june.rentals_per_customer_june.sum() - rentals_may_and_june.rentals_per_customer_may.sum() > 0:
    print('Rental difference between June and May: ', rentals_may_and_june.rentals_per_customer_june.sum() - rentals_may_and_june.rentals_per_customer_may.sum())
    print('Customers borrowed more films in June than May')
else:
    print('Rental difference between June and May: ', rentals_may_and_june.rentals_per_customer_june.sum() - rentals_may_and_june.rentals_per_customer_may.sum())
    print ('Customers did not borrowed more films in June than May')
```

```python
import numpy as np
```

```python
# function to check if each customer borrowed more films in june comparing with may

rentals_may_and_june['rented_more_in_june'] = np.where(rentals_may_and_june['rentals_per_customer_june'] - rentals_may_and_june['rentals_per_customer_may'] > 0, True, False)
display(rentals_may_and_june)
print('Customers that rented more in June: ')
print(rentals_may_and_june.rented_more_in_june.value_counts())
```
